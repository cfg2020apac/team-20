# User Profile

```python
import pandas as pd
import numpy as np
import names
```

```python
## Load data
user_f = pd.read_excel('handson_data/2. Dummy Volunteer Profile (updated) .xls', index_col=None)
user_f.head()
```

```python
## Modify column names
user_f_columns = ['first_name','last_name','interest_list','gender','employment_status','mobile','connection_id','email','attend_status','hours_served_opportunity','hours_served_all','opportunity_name','opportunity_org','opportunity_type','employer','contact_id']
user_f.columns = user_f_columns
user_f.head()
```

```python
## Generate Names
first_names = []
last_names = []
for i in range(len(user_f)):
    gender = user_f.gender.loc[i].lower()
    first_name = names.get_first_name(gender=gender)
    last_name = names.get_last_name()
    first_names.append(first_name)
    last_names.append(last_name)

user_f.first_name = first_names
user_f.last_name = last_names
user_f.head()
```

```python
## Reform the interests to list
user_f.interest_list = user_f.interest_list.str.replace('?','')
user_f.interest_list = user_f.interest_list.str.split(';')
user_f.head()
```

```python
## Generate dummy email address
user_f.email = user_f.first_name.str.lower() + '.' + user_f.last_name.str.lower() + '@xmail.com'
user_f.head()
```

```python
## Remove weird characters from opportunity_org
user_f.opportunity_org = user_f.opportunity_org.str.replace('?','')
user_f.head()
```

```python
user_f.to_csv('handson_data/dummy_volunteer_profile.csv')
```

# HandsOn Program
